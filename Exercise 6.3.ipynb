# 7. Choropleth map

```python
print(coral_bleach_2010_2019.head())
```

```python
# creating a subset:

data_to_plot = coral_bleach_2010_2019[['Country_Name', 'Turbidity']]
```

```python
# Standarizing country names so they can be matched:

country_name_replacements = {
    'The Bahamas': 'Bahamas',
    'United States of America': 'United States',
    'United Republic of Tanzania': 'Tanzania'
}

data_to_plot['Country_Name'] = data_to_plot['Country_Name'].replace(country_name_replacements)
```

```python
for feature in country_coral['features']:
    country_name = feature['properties']['name']
    # Check if the country name needs to be replaced
    if country_name in country_name_replacements:
        feature['properties']['name'] = country_name_replacements[country_name]
```

```python
print(data_to_plot['Turbidity'].head())
print(data_to_plot['Turbidity'].dtypes)
print(data_to_plot['Turbidity'].isnull().sum())
```

```python
data_to_plot['Turbidity'] = data_to_plot['Turbidity'].str.replace(',', '.')
```

```python
data_to_plot['Turbidity'] = pd.to_numeric(data_to_plot['Turbidity'], errors='coerce')
```

```python
map = folium.Map(location=[0, 0], zoom_start=2)


folium.Choropleth(
    geo_data=country_coral,  
    data=data_to_plot,
    columns=['Country_Name', 'Turbidity'],
    key_on='feature.properties.name',
    fill_color='YlOrBr',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Turbidity Levels',
).add_to(map)


folium.LayerControl().add_to(map)


map
```
