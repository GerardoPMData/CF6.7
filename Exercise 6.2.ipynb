# 3. Create a correlation matrix heatmap (colored)

```python
# Select only the numeric columns for correlation
numerical_cols = ['Cyclone_Frequency', 'Depth_m', 'Distance_to_Shore', 
                  'Turbidity', 'Temperature_Maximum', 'Percent_Bleaching', 'Temperature_Mean']
```

```python
# Create a correlation matrix
corr_matrix = coral_bleach_2010_2019[numerical_cols].corr()
```

```python
# Create the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True)

plt.show()
```

```python
coral_bleach_2010_2019[['Turbidity', 'Distance_to_Shore']].describe()
```

# 3.1 Results of the correlation matrix:

```python
# Strong positive correlation (values close to 1) indicates that as one variable increases, the other tends to increase as well. For example, "Temperature_Maximum" and "Percent_Bleaching" have a correlation of 0.39, suggesting a moderate positive relationship.
# Negative correlations (values closer to -1) indicate that as one variable increases, the other tends to decrease. For example, "Cyclone_Frequency" and "Percent_Bleaching" have a slight negative correlation of -0.06.
# Low or no correlation (values close to 0) indicates no significant relationship between variables, like "Turbidity", however, the dataframe seems to be inconsistant, as Turbidity contains indeed numeric values but theyre not tecognized as such, and thus, no correlation can be calculated
```
